## Solution 1

```python
def solution_1(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.maximize_window()
    browser.get(URL)
    time.sleep(2)
    ## storing all table rows element in a list
    tbl_rows=browser.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr")[:30]
    rank=[]
    name=[]
    artist=[]
    views=[]
    date=[]
    ## traversing each row inside table one by one
    for i in tbl_rows:
        ## collecting data from data inside the rows like rank,artist etc
        try: rank.append(i.find_element_by_xpath(".//td[1]").text)
        except:rank.append('Not found')
        try:  name.append(i.find_element_by_xpath(".//td[2]").text)
        except: name.append('Not found')
        try:artist.append(i.find_element_by_xpath(".//td[3]").text)
        except: artist.append('Not found')
        try:views.append(i.find_element_by_xpath(".//td[4]").text)
        except: views.append('Not found')
        try:date.append(i.find_element_by_xpath(".//td[5]").text)
        except: date.append('Not found')
    df=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Views (billions)':views,'Upload Date':date})
    return df
```

## solution_2

```python
def solution_2(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.maximize_window()
    browser.get(URL)
    ## getting URL for international fixture page
    fixture_page_url=browser.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']/div/ul/li/a").get_attribute('href')
    ## navigating to international fixture page
    browser.get(fixture_page_url)
    ## collecting all anchor tags available on page
    anchors=browser.find_elements_by_xpath("//div[@class='event-list__list']/div/a")
    ## declaring empty lists to collect data
    title=[]
    series=[]
    place=[]
    time=[]
    date=[]
    ## traversing all anchor tags one by one and collecting required data
    for i in anchors:
        ## collecting match title
        try:title.append(i.find_element_by_xpath(".//div[@class='fixture__info u-skewed']/div[@class='fixture__description u-unskewed-text']/p[@class='fixture__additional-info']/strong[@class='fixture__name fixture__name--with-margin']").text)
        ## if exception occurs appending NA
        except:title.append('NA')
        # similarly collecting series
        try:series.append(i.find_element_by_xpath(".//div[@class='fixture__info u-skewed']/div[@class='fixture__format-strip']/span[@class='u-unskewed-text fixture__tournament-label u-truncated']").text)    
        except:series.append('NA')
        ## collecting place
        try:place.append(i.find_element_by_xpath(".//div[@class='fixture__info u-skewed']/div[@class='fixture__description u-unskewed-text']/p[@class='fixture__additional-info']/span").text)
        except:place.append('NA')
        ## declaring empty string 
        dt=''
        try:
            ## appending date and month in string dt
            dt=dt+(i.find_element_by_xpath(".//div[@class='fixture__full-date']/span[@class='fixture__date']").text)
            dt=dt+(i.find_element_by_xpath(".//div[@class='fixture__full-date']/div[@class='fixture__date-details']/span[@class='fixture__month']").text)
            ## appending dt in date list
            date.append(dt)
        except:date.append('NA')
        ## collecting Time
        try:time.append(i.find_element_by_xpath(".//div[@class='fixture__full-date']/div[@class='fixture__date-details']/span[@class='fixture__time']").text)
        except: time.append('NA')
    df=pd.DataFrame({'Match Title':title,'Series':series,'Place':place,'Date':date,'Time':time})
    return df
```

## Solution 3

```python
def solution_3(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.maximize_window()
    browser.get(URL)
    ## searching for selenium out of all available anchor tags (Title on page) and once selenium is found breaking the loop
    for k in browser.find_elements_by_xpath("//li[@class='fa fa-chevron-circle-right']/a"):
        if(k.text=='Selenium'):
            browser.get(k.get_attribute('href'))
            break
    ## Now storing all rows of table in a list
    rows=browser.find_elements_by_xpath("//table[@class='table']/tbody/tr")
    ## traversing all rows till 'Selenium Exception Handling (Common Exceptions List)' is found and once found braking the loop after navigating to that page
    for td in rows:
        if(td.find_element_by_xpath(".//td[2]").text=='Selenium Exception Handling (Common Exceptions List)'):
            url=td.find_element_by_xpath(".//td[1]/a").get_attribute('href')
            browser.get(url)
            break
    ## storing all rows available on page containing name and description
    data_rows=browser.find_elements_by_xpath("//table[@class='table table-striped']/tbody/tr")[1:]
    ## declaring empty lists
    name=[]
    desc=[]
    ## traversing all data rows and storing data and description
    for data in data_rows:
        name.append(data.find_element_by_xpath(".//td[1]").text)
        desc.append(data.find_element_by_xpath(".//td[2]").text) 
    df=pd.DataFrame({'Name':name,'Description':desc})
    return df
```

## Solution 4

```python
def solution_4(URL):

    browser=webdriver.Chrome('chromedriver.exe')
    browser.maximize_window()
    browser.get(URL)
    ## getting the element containing 'economy india' option
    economy_india=browser.find_element_by_xpath("//div[@class='navbar']/div[2]/div[@class='dropdown-content']/a[3]")
    ## navigating to economy india page
    browser.get(economy_india.get_attribute('href'))
    ## getting list of all avilable anchor tags
    list_tags=browser.find_elements_by_xpath("//ul/li/a")
    ## taraversing all anchors and stopping the loop once 'GDP of Indian states' is found
    for i in list_tags:
        if(i.text=='» GDP of Indian states'):
            browser.get(i.get_attribute('href'))
            break
    ## getting all tables available on page
    gdp_table=browser.find_elements_by_xpath("//table[@class='display dataTable']")
    ## traversing all table till GDP table is found
    for i in gdp_table:
        if i.find_element_by_xpath(".//caption").text=='GSDP':
            ## if table is GDP then storting all rows inside it
            gdp_rows=i.find_elements_by_xpath(".//tbody/tr")
            break
    ## declaring empty lists
    rank=[]
    state=[]
    gsdp_1819=[]
    gsdp_1920=[]
    share=[]
    gdp=[]
    ## traversing all table rows one by one and collecting required data
    for k in gdp_rows:
        ## storing rank
        try:rank.append(k.find_element_by_xpath(".//td[1]").text)
        except:rank.append('NA')
        try:state.append(k.find_element_by_xpath(".//td[2]").text)
        except:state.append('NA')
        try:gsdp_1819.append(k.find_element_by_xpath(".//td[4]").text)
        except:gsdp_1819.append('NA')
        try:gsdp_1920.append(k.find_element_by_xpath(".//td[3]").text)
        except:gsdp_1920.append('NA')
        try:share.append(k.find_element_by_xpath(".//td[5]").text)
        except: share.append('NA')
        try: gdp.append(k.find_element_by_xpath(".//td[6]").text)
        except: gdp.append('NA')
    df=pd.DataFrame({'Rank':rank,'State':state,'GSDP(18-19)':gsdp_1819,'GSDP(19-20)':gsdp_1920,'Share':share,'GDP ($ billion)':gdp}) 
    return df
```

## solution 5

```python

def solution_5(URL):

    browser=webdriver.Chrome('chromedriver.exe')
    browser.maximize_window()
    browser.get(URL)
    ## searching for treding URL
    trending_url=browser.find_element_by_xpath("//ul[@class='list-style-none mb-3']/li[3]/a")
    ## navigating to trending URL
    browser.get(trending_url.get_attribute('href'))
    ## getting all repository URL
    repository_href=[i.get_attribute('href') for i in browser.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")]
    ## getting names of all respository
    names=[i.text for i in browser.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")]
    ## declaring empty list
    desc=[]
    contri_count=[]
    language=[]
    ## navigating inside all repositories one by one
    for k in repository_href:
        ## making temporary variable
        temp=False
        ## refreshing browser to avoid stale element exception
        browser.refresh()
        try:
            ## navigating  to each repository
            browser.get(k)  
            ## storing description
            try:desc.append(browser.find_element_by_xpath("//div[@class='BorderGrid-row hide-sm hide-md']/div[1]/p[@class='f4 mt-3']").text)
            except:desc.append('NA')
            ## storing contribution count
            try:
                for i in browser.find_elements_by_xpath("//div[@class='BorderGrid-row']/div[@class='BorderGrid-cell']/h2[@class='h4 mb-3']"):
                    if (i.text.split(' ')[0]=='Contributors'):
                        temp=True
                        temp=i.text.split(' ')[1]
                        contri_count.append(temp)
                if(temp==False): contri_count.append('NA')
            except:contri_count.append('NA')
            ## storing languages
            try:
                lang=browser.find_elements_by_xpath("//li[@class='d-inline']")
                strng=''
                for i in lang: strng=strng+i.text.replace('\n','-')+','
                language.append(strng)
            except: language.append('NA')
        ## handling stale element exception in case it occurs
        except StaleElementReferenceException:
            browser.back()
            browser.get(k)

    df=pd.DataFrame({'Repository Title':names,'repository Description':desc,'Contributors count':contri_count,'Languages':language})
    return df
```

## Solution 6

```python
def solution_6(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.maximize_window()
    browser.get(URL)

    ## getting the href of hot_100 button
    hot_100=browser.find_element_by_xpath("//ul[@class='header__submenu__list']/li[1]/a").get_attribute('href')
    ## navigatin to hot_100 page
    browser.get(hot_100)
    ## storing all_songs
    all_songs_tags=browser.find_elements_by_xpath("//li[@class='chart-list__element display--flex']/button[@class='chart-element__wrapper display--flex flex--grow sort--this-week']")
    ## declaring empty lists
    name=[]
    artist=[]
    last_week=[]
    peak_rank=[]
    week_on_board=[]
    ## traversing all_songs_tags
    for i in all_songs_tags:
        ## collecting name
        name.append(i.find_element_by_xpath(".//span[@class='chart-element__information']/span[@class='chart-element__information__song text--truncate color--primary']").text)
        ## collecting artist name
        artist.append(i.find_element_by_xpath(".//span[@class='chart-element__information__artist text--truncate color--secondary']").text)
        ## collecting last week rating
        last_week.append(i.find_element_by_xpath(".//div[@class='chart-element__meta text--center color--secondary text--last']").text)
        ## collecting peak rank
        peak_rank.append(i.find_element_by_xpath(".//div[@class='chart-element__meta text--center color--secondary text--peak']").text)
        ## collecting weeks on board
        week_on_board.append(i.find_element_by_xpath(".//div[@class='chart-element__meta text--center color--secondary text--week']").text)
    ## forming dataframe
    df=pd.DataFrame({'Song name':name,'Artist name':artist,'Last week rank':last_week,'Peak rank':peak_rank,'Weeks on board':week_on_board})
    return df
```

## Solution 7

```python
def solution_7(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.maximize_window()
    browser.get(URL)
    ## getting recruiters url
    recruiters=browser.find_element_by_xpath("//ul[@class='midSec menu']/li[2]/a").get_attribute('href')
    ## navigating to recruiters page
    browser.get(recruiters)
    ## finding serach bar and entering 'Data science ' in it
    browser.find_element_by_xpath("//input[@class='sugInp']").send_keys('Data Science')
    ## finding search button and clicking it
    browser.find_element_by_xpath("//button[@class='fl qsbSrch blueBtn']").click()

    ## getting all left hand blocks
    sections_left=browser.find_elements_by_xpath("//div[@class='recSec fl fadeInUp']")
    ## getting all right hand blocks
    sections_right=browser.find_elements_by_xpath("//div[@class='recSec fr fadeInUp']")
    ## declaring empty lists
    name=[]
    desig=[]
    company=[]
    place=[]
    skills=[]
    ## combining right hand and left hand blocks
    total_sections=sections_left+sections_right
    ## traversing all blocks one by one and collecting required data
    for i in total_sections:
        try:name.append(i.find_element_by_xpath(".//p[@class='highlightable']/a/span").text)
        except: name.append('NA')
        try:desig.append(i.find_element_by_xpath(".//p[@class='highlightable']/span[@class='ellipsis clr']").text)
        except:desig.append('NA')
        try: company.append(i.find_element_by_xpath(".//p[@class='highlightable']/a[2]").text)
        except: company.append('NA')
        try: place.append(i.find_element_by_xpath(".//p[@class='highlightable']/span[2]").text)
        except: place.append('NA')
        try: skills.append(i.find_element_by_xpath(".//div[@class='hireSec highlightable']").text)
        except: place.append('NA')
    ##framing dataframe
    df=pd.DataFrame({'Name':name,'Designation':desig,'Company':company,'Skills':skills,'Location':place})
    return df
```

## solution 9

```python
def solution_9(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.maximize_window()
    browser.get(URL)
    ## getting all available movie blocks on the page
    all_list=browser.find_elements_by_xpath("//div[@class='lister-item mode-detail']")
    ## declaring empty lists
    name=[]
    y_span=[]
    genre=[]
    runtime=[]
    rating=[]
    votes=[]
    ## traversing all block one by one and collecting data
    for i in all_list:
        ## getting name of movie
        try:name.append(i.find_element_by_xpath(".//h3[@class='lister-item-header']/a").text)
        except: name.append('NA')
        ## getting year span
        try:y_span.append(i.find_element_by_xpath(".//h3[@class='lister-item-header']/span[@class='lister-item-year text-muted unbold']").text)
        except: y_span.append('NA')
        ## getting Genre
        try:genre.append(i.find_element_by_xpath(".//p[@class='text-muted text-small']/span[@class='genre']").text)
        except: genre.append('NA')
        ## getting runtime
        try:runtime.append(i.find_element_by_xpath(".//p[@class='text-muted text-small']/span[@class='runtime']").text)
        except: runtime.append('NA')
        ## getting Rating
        try:rating.append(i.find_element_by_xpath(".//div[@class='ipl-rating-widget']/div[@class='ipl-rating-star small']").text)
        except: rating.append('NA')
        ## getting votes
        try:votes.append(i.find_element_by_xpath(".//p[4]/span[2]").text)
        except: votes.append('NA')
    df=pd.DataFrame({'Name':name,'Year span':y_span,'Genre':genre,'Runtime':runtime,'Ratings':rating,'Votes':votes})
    return df
```

## solution 10

```python
def solution_10(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.maximize_window()
    browser.get(URL)



    ## getting all anchors tags
    all_dataset=browser.find_elements_by_xpath("//span[@class='whitetext']/a")

    ## finding all dataset anchor tag and navigating to that url
    for i in all_dataset:
        if i.text  == 'View ALL Data Sets':
            browser.get(i.get_attribute('href'))
            break

    ## storing all table rows
    table_rows=browser.find_elements_by_xpath("//table[2]/tbody[1]/tr[1]/td[2]/table[2]/tbody/tr")[1:]
    ## declaring all empty lists
    name=[]
    data_type=[]
    task=[]
    Attribute_type=[]
    No_of_instances=[]
    No_of_attribute=[]
    year=[]



    ## traversing all table rows one by one and collecting data
    for row in table_rows:
        try: name.append(row.find_element_by_xpath(".//td[1]").text)
        except: name.append('NA')
        try: data_type.append(row.find_element_by_xpath(".//td[2]").text)
        except: data_type.append('NA')
        try: task.append(row.find_element_by_xpath(".//td[3]").text)
        except: task.append('NA')
        try: Attribute_type.append(row.find_element_by_xpath(".//td[4]").text)
        except: Attribute_type.append('NA')
        try: No_of_instances.append(row.find_element_by_xpath(".//td[5]").text)
        except: No_of_instances.append('NA')
        try: No_of_attribute.append(row.find_element_by_xpath(".//td[6]").text)
        except: No_of_attribute.append('NA')
        try: year.append(row.find_element_by_xpath(".//td[7]").text)
        except: year.append('NA')
    ## framing a dataframe
    df=pd.DataFrame({'Dataset name':name,'Data type':data_type,'Task':task,'Attribute type':Attribute_type,'No_of_instances':No_of_instances,'No_of_attribute':No_of_attribute,'year':year})
    return df
```
