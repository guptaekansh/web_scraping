```python
################ importing necessary Libraries #######################
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

## Solution of question 1

```python
##################################### Solution to question 1 #####################################

def solution_1(URL):
    page_content =  requests.get(URL)
    html_parsed_content = BeautifulSoup(page_content.content,'html.parser')
    all_headers=html_parsed_content.find_all('h2')
    all_headers_text=[]
    for i in all_headers:
        all_headers_text.append(i.get_text())
    Headers=pd.DataFrame(all_headers_text,columns=['Header_name'])
    return Headers
    
```

```python
solution_1('https://en.wikipedia.org/wiki/Main_Page')
```

## Solution of question 2

```python
##################################### Solution to question 2 and 3 #####################################
import re
#URL='https://www.imdb.com/list/ls091520106/'
def solution_2and3(URL):
    page_content =  requests.get(URL).text
    html_parsed_content = BeautifulSoup(page_content,'html.parser')
    html_parsed_content
    Name=[]
    Year=[]
    Rating=[]
    ratings=html_parsed_content.find_all('div',class_='ipl-rating-star small')
    lst=html_parsed_content.find_all('h3',class_='lister-item-header')
    for i,k in zip(lst,ratings):
        name=i.find('a').get_text()
        Name.append(name)
        year=i.find('span',class_='lister-item-year text-muted unbold').get_text()
        Year.append(re.sub("[()]","",year))
        r=k.findChildren('span',class_="ipl-rating-star__rating")[0].get_text()
        Rating.append(r)
    Movies=pd.DataFrame({'Name':Name,'IMDB Rating':Rating,'Release year':Year})
    Movies.to_csv('Movies.csv')
    return Movies
```

## Solution of question 4

```python
########################################### solution to question 4 #############################################
#URL="https://bookpage.com/reviews"
def solution_4(URL):
    content=requests.get(URL).text
    html_parsed_content=BeautifulSoup(content,'html.parser')
    Book_names=[i.get_text().replace("\n","").strip() for i in html_parsed_content.find_all('h4',{'class':'italic'},limit=5)]
    Author_name=[i.get_text().replace("\n","").strip() for i in html_parsed_content.find_all('p',{'class':'sans bold'},limit=5)]
    Genre=[i.get_text().replace("\n","").strip() for i in html_parsed_content.find_all('p',{'class':'genre-links hidden-phone'},limit=5)]
    Review=[i.get_text().replace("\n","").strip() for i in html_parsed_content.find_all('p',{'class':'excerpt'},limit=5)]
    Books=pd.DataFrame({'Name':Book_names,'Auhtor':Author_name,'Genre':Genre,'Review':Review})
    return Books
```

### Solution to quesiton 5

```python
################################# Solution to question 5 part 1 #########################################
#URL="https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
def solution_5and6_part_1(URL):
    content=requests.get(URL).text
    html_parsed_content=BeautifulSoup(content,'html.parser')
    table= html_parsed_content.find('table',class_='table')
    top10_ODIteams_mens_names=[i.get_text() for i in table.find_all('span',class_='u-hide-phablet',limit=10)]
    top10_ODIteams_mens_matches=[]
    top10_ODIteams_mens_points=[]
    top10_ODIteams_mens_ratings=[]

    for i in table.find_all('tr', class_="table-body",limit=9):
        m,p=i.findChildren('td',class_="table-body__cell u-center-text")
        top10_ODIteams_mens_matches.append(m.get_text())
        top10_ODIteams_mens_points.append(p.get_text())
        rt=i.findChildren('td',class_="table-body__cell u-text-right rating")
        top10_ODIteams_mens_ratings.append(rt[0].get_text())

    top10_ODIteams_mens_matches.insert(0,table.find('td',class_='rankings-block__banner--matches').get_text())
    top10_ODIteams_mens_points.insert(0,table.find('td',class_='rankings-block__banner--points').get_text())
    top10_ODIteams_mens_ratings.insert(0,table.find('td',class_='rankings-block__banner--rating u-text-right').get_text().replace("\n",'').replace(" ",""))
    top10_ODIteams_mens=pd.DataFrame({'Team':top10_ODIteams_mens_names,'Macthes':top10_ODIteams_mens_matches,'Points':top10_ODIteams_mens_points,'Ratings':top10_ODIteams_mens_ratings})
    return top10_ODIteams_mens
```

```python
################################ solution to question 5 part 2 and 3 ##############################
def solution_5and6_part_2and3(URL):
    content=requests.get(URL).text
    html_parsed_text=BeautifulSoup(content,'html.parser')
    table=html_parsed_text.find_all('tr',class_='table-body',limit=9)
    Names=[i.findChildren('a')[0].get_text() for i in table]
    Team=[i.findChildren('span',class_="table-body__logo-text")[0].get_text() for i in table]
    Rating=[i.findChildren('td',class_="table-body__cell rating")[0].get_text() for i in table]
    Names.insert(0,html_parsed_text.find('div',class_="rankings-block__banner--name-large").get_text())
    Team.insert(0,html_parsed_text.find('div',class_="rankings-block__banner--nationality").get_text().replace("\n",""))
    Rating.insert(0,html_parsed_text.find('div',class_="rankings-block__banner--rating").get_text().replace("\n",""))
    return pd.DataFrame({'Name':Names,'Team':Team,'Rating':Rating})
```

## Solution to problem 7

```python
#URL='https://www.amazon.in/Mobile-Phone-Under-20000-Rupees/s?k=Mobile+Phone+Under+20000+Rupees'
def solution_7(URL):
    driver=webdriver.Chrome('chromedriver.exe')
    page=driver.get(URL)
    phones=pd.DataFrame()
    pge=1
    cnt=0
    while(cnt<=2):
        Names=[i.text for i in driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")]
        Price=[i.text for i in driver.find_elements_by_xpath("//span[@class='a-price-whole']")]
        Image_URL=[i.get_attribute('src') for i in driver.find_elements_by_xpath("//div[@class='a-section aok-relative s-image-fixed-height']/img")]
        Rating=[i.get_attribute('aria-label') for i in driver.find_elements_by_xpath("//div[@class='a-row a-size-small']/span[1]")]
        if((len(Names)==len(Price)) & (len(Price)==len(Image_URL)) & (len(Image_URL)==len(Rating))):
            df=pd.DataFrame({'Phone_name':Names,'Price':Price,'Image_URL':'Image_URL','Avg_Rating':Rating})
            phones=pd.concat([df,phones])
            cnt=cnt+1
        try:
            button=driver.find_element_by_xpath("//li[@class='a-last']/a")
            button.click()
            #time.wait(3)
            driver.refresh()
            driver.implicitly_wait(5)
            #pge=pge+1
        except:
            break
    return phones
```

## solution to problem 9

```python
##URL=https://www.monsterindia.com/srp/results?query=software%20developer&searchId=43efda09-8cdf-4757-90ad-15420996f0bc
def solution_9and10(URL):
    driver_=webdriver.Chrome('chromedriver.exe')
    page=driver_.get(URL)
    monster_jobs=pd.DataFrame()
    for k in range(5):
        job_name=[i.text.replace("\n","") for i in driver_.find_elements_by_xpath("//div[@class='job-tittle']/h3")]
        Company=[i.text for i in driver_.find_elements_by_xpath("//span[@class='company-name']")]
        Place=[i.text for i in driver_.find_elements_by_xpath("//div[@class='col-xxs-12 col-sm-5 text-ellipsis']/span[@class='loc']/small")]
        Experience=[i.text for i in driver_.find_elements_by_xpath("//div[@class='exp col-xxs-12 col-sm-3 text-ellipsis']/span/small")]
        if((len(job_name)==len(Place)) & (len(Place)==len(Experience))):
            df=pd.DataFrame({'JOB_NAME':job_name,'Company':Company,'Place':Place,'Experience':Experience})
            monster_jobs=pd.concat([df,monster_jobs])
        else:
            print('Lengths of lists do not match')
            break
        button=driver_.find_element_by_xpath("//button[@class='btn-next-prev btn-next']")
        button.click()
        driver_.refresh()
        driver_.implicitly_wait(5)
    return monster_jobs
```
