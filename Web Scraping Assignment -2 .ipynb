```python

def solution_4(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.get(URL)
    browser.find_element_by_id('sc.keyword').send_keys('Data Scientist')

    browser.find_element_by_id('sc.location').send_keys(Keys.CONTROL+"a")
    browser.find_element_by_id('sc.location').send_keys(Keys.DELETE)
    time.sleep(1)
    browser.find_element_by_id('sc.location').send_keys('Noida')
    browser.find_element_by_xpath("//button[@class='gd-ui-button ml-std col-auto SearchStyles__newSearchButton css-iixdfr']").click()
    browser.refresh()
    WebDriverWait(browser,20).until(EC.presence_of_element_located((By.ID,'onetrust-consent-sdk')))

    jobs=[]
    jobs.append(browser.find_element_by_xpath("//li[@class='jl react-job-listing gdGrid selected']"))
    jobs.extend(browser.find_elements_by_xpath("//li[@class='jl react-job-listing gdGrid ']")[0:9])

    ## extracting days 
    days=[i.text for i in browser.find_elements_by_xpath("//div[@class='d-flex align-items-end pl-std css-mi55ob']")[0:10]]

    ## extracting company name and rating
    company=[]
    rating=[]
    for i in jobs:
        i.click()
        #WebDriverWait(browser,20).until(EC.presence_of_element_located((By.XPATH,"//div[@class='css-87uc0g e1tk4kwz1']")))
        time.sleep(5)
        k=browser.find_element_by_xpath("//div[@class='css-87uc0g e1tk4kwz1']").text
        company.append(k.split("\n")[0])
        try:
            rating.append(browser.find_element_by_xpath("//span[@class='css-1m5m32b e1tk4kwz2']").text)
        except:
            rating.append('Not Available')
    if((len(days)==len(company)) & (len(company)==len(rating))):
        df=pd.DataFrame({'Company_name':company,'Days since job posted':days,'Rating':rating})
        return df
    else:
        print('Length of lists do not match')
```

```python
def solution_5(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.get(URL)
    ## searching and entering data in keyword field
    browser.find_element_by_id('KeywordSearch').send_keys('Data Scientist')
    ## finding location field element
    location=browser.find_element_by_id('LocationSearch')
    ## selecting all text in location field
    location.send_keys(Keys.CONTROL+"a")
    ## removing al text 
    location.send_keys(Keys.DELETE)
    time.sleep(1)
    ## entering Noida as location 
    location.send_keys('Noida')
    ## loacting and clicing search button
    browser.find_element_by_xpath("//button[@id='HeroSearchButton']").click()
    WebDriverWait(browser,20).until(EC.presence_of_element_located((By.ID,'SalaryFilters')))
    ##extarcting required data
    company_name=[i.text for i in browser.find_elements_by_xpath("//p[@class='m-0 ']")[0:10]]
    no_of_salaries=[i.text for i in browser.find_elements_by_xpath("//p[@class='css-1uyte9r css-1kuy7z7 m-0 ']")[0:10]]
    avg_salary=[i.text for i in browser.find_elements_by_xpath("//div[@class='col-2 d-none d-md-flex flex-row justify-content-end']/strong")[0:10]]
    min_salary=[i.text.split('\n')[0] for i in browser.find_elements_by_xpath("//div[@class='common__RangeBarStyle__values common__flex__justifySpaceBetween common__flex__container ']")[0:10]]
    max_salary=[i.text.split('\n')[1] for i in browser.find_elements_by_xpath("//div[@class='common__RangeBarStyle__values common__flex__justifySpaceBetween common__flex__container ']")[0:10]]
    if((len(no_of_salaries)==len(company_name)) & (len(company_name)==len(avg_salary)) & (len(avg_salary)==len(min_salary))):
        df=pd.DataFrame({'Company_name':company_name,'No_of_salaries':no_of_salaries,'Average_salary':avg_salary,'Min_salary':min_salary,'Max_salary':max_salary})
        return df
    else:
        print('Lenght od lists do not match')
```

## Solution 6

```python
def solution_6(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.get(URL)
    browser.find_element_by_xpath("//input[@class='_3704LK']").send_keys('sunglasses') ## entering sunglasses in search box
    browser.find_element_by_xpath("//button[@class='L0Z3Pu']").click() ## clicking search icon
    glasses=pd.DataFrame()
    while(glasses.shape[0]<100):
        WebDriverWait(browser,20).until(EC.presence_of_element_located((By.XPATH,"//div[@class='_3I9_wc']")))
        ## getting all required data
        brand=[i.text for i in browser.find_elements_by_xpath("//div[@class='_2WkVRV']")]
        prod_desc=[i.text for i in browser.find_elements_by_xpath("//div[@class='_2B099V']/a[1]")]
        price=[i.text for i in browser.find_elements_by_xpath("//div[@class='_30jeq3']")]
        discount=[i.text for i in browser.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span")]
        ## checking length of lists created
        if((len(brand)==len(prod_desc)) & (len(prod_desc)==len(price)) & (len(price)==len(discount))):
            df1=pd.DataFrame({'Brand':brand,'product_description':prod_desc,'Price':price,'Discount':discount})
            glasses=glasses.append(df1,ignore_index=True)
        else:
            print('Length of lists do not match')
        try:
            next_page=browser.find_element_by_xpath("//a[@class='_1LKTO3'][2]").get_attribute('href')
        except:
            next_page=browser.find_element_by_xpath("//a[@class='_1LKTO3']").get_attribute('href')
        browser.get(next_page)
        browser.refresh()
    browser.close()
    return glasses.iloc[0:100,:]
```

```python
def solution_7(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.get(URL)
    
    ## finding all review element 
    all_review=browser.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']")
    parent=all_review.find_element_by_xpath('..')
    ## navigating to all review page
    browser.get(parent.get_attribute('href'))
    
    ## creating empty list
    rating=[]
    review_summary=[]
    full_review=[]
    ## getting data in list
    while(len(rating)<=100):
        for i in browser.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]"):
            rating.append(i.text)
        for i in browser.find_elements_by_xpath("//p[@class='_2-N8zT']"):
            review_summary.append(i.text)
        for i in browser.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div"):
            full_review.append(i.text)
        ## storing next page element
        try:    
            next_page=browser.find_element_by_xpath("//a[@class='_1LKTO3'][2]").get_attribute('href')
        except:
            next_page=browser.find_element_by_xpath("//a[@class='_1LKTO3']").get_attribute('href')
        ## moving to next page
        browser.get(next_page)
        browser.refresh()
        WebDriverWait(browser,20).until(EC.presence_of_element_located((By.CLASS_NAME,'_2d4LTz')))
        
    if((len(rating)==len(review_summary)) & (len(review_summary)==len(full_review))):
        temp_df=pd.DataFrame({'Rating':rating,'review_summary':review_summary,'full_review':full_review})
    else:
        print('Length of lists do not match')
    browser.close()
    return temp_df.iloc[0:100,:]
```

```python
def solution_8(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.get(URL)
    browser.find_element_by_xpath("//input[@class='_3704LK']").send_keys('sneakers') ## entering sunglasses in search box
    browser.find_element_by_xpath("//input[@class='_3704LK']").send_keys(Keys.ENTER) ## pressing Enter key
    WebDriverWait(browser,60).until(EC.presence_of_element_located((By.CLASS_NAME,'_3I9_wc')))
    brand=[]
    prod_desc=[]
    discount=[]
    price=[]
    while(len(brand)<=100):
        ## getting required data
        for i in browser.find_elements_by_xpath("//div[@class='_2WkVRV']"):brand.append(i.text)
        for i in browser.find_elements_by_xpath("//div[@class='_2B099V']/a[1]"):prod_desc.append(i.text)
        for i in browser.find_elements_by_xpath("//div[@class='_30jeq3']"):price.append(i.text)
        for i in browser.find_elements_by_xpath("//a[@class='_3bPFwb']/div[@class='_25b18c']"):discount.append(i.text)
        s=[]
        for i in discount:
            if '%' in i:s.append(i[-7:])
            else:s.append('Null')
        try:    
            next_page=browser.find_element_by_xpath("//a[@class='_1LKTO3'][2]").get_attribute('href')
        except:
            next_page=browser.find_element_by_xpath("//a[@class='_1LKTO3']").get_attribute('href')
        browser.get(next_page)
        browser.refresh()
        WebDriverWait(browser,20).until(EC.presence_of_element_located((By.CLASS_NAME,'_3I9_wc')))
        print(len(brand),len(prod_desc),len(price),len(s))
    if((len(brand)==len(prod_desc)) & (len(prod_desc)==len(price)) & (len(price)==len(s))):
        df=pd.DataFrame({'Brand':brand,'product_description':prod_desc,'Price':price,'Discount':s})
    else:
        print('Length of lists do not match')
    browser.close()
    return df.iloc[0:100,:]
```

```python
def solution_9(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.get(URL)
    WebDriverWait(browser,20).until(EC.presence_of_element_located((By.CLASS_NAME,"vertical-filters-header")))
    ##checking price filter
    browser.find_element_by_xpath("//ul[@class='price-list']/li[2]/label").click()
    browser.refresh()
    ## checking color filter
    browser.find_element_by_xpath("//li[@class='colour-listItem'][1]").click()
    browser.refresh()
    ## initializing lists
    brand=[]
    short_desc=[]
    price=[]
    pr=[]
    while(len(brand)<=100):
        ## getting all required data
        for i in browser.find_elements_by_xpath("//h3[@class='product-brand']"):brand.append(i.text)
        for i in browser.find_elements_by_xpath("//h4[@class='product-product']"):short_desc.append(i.text)
        for i in browser.find_elements_by_xpath("//div[@class='product-price']/span[1]"):price.append(i.text)
        ## navigating to next page
        browser.find_element_by_xpath("//li[@class='pagination-next']").click()
        browser.refresh()
    ## resolving price list to get correct price   
    for i in price:pr.append('Rs. '+(i.split(' ')[1].replace('Rs.','')))
    ## comparing the lenght of lists   
    if((len(brand)==len(short_desc)) & (len(short_desc)==len(pr)) ):
        df=pd.DataFrame({'Brand':brand,'Short_Description':short_desc,'Price':pr})
    else:
        print('Lenght of lists do not match')
    browser.close()
    return df.iloc[0:100,:]
```

```python
from bs4 import BeautifulSoup
import requests

def solution_10(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.get(URL)
    ## finding search box and entring 'Laptop' in the search field
    browser.find_element_by_xpath("//div[@class='nav-search-field ']/input").send_keys('Laptop')
    ## clikcing the search button
    browser.find_element_by_id("nav-search-submit-button").click()
    ## applying filter 1
    filter1=browser.find_element_by_xpath("//li[@id='p_n_feature_thirteen_browse-bin/12598163031']/span[@class='a-list-item']/a[@class='a-link-normal s-navigation-item']")
    browser.get(filter1.get_attribute('href'))
    time.sleep(5)
    ## appying filter 2
    filter2=browser.find_element_by_xpath("//li[@id='p_n_feature_thirteen_browse-bin/16757432031']/span[@class='a-list-item']/a[@class='a-link-normal s-navigation-item']")
    browser.get(filter2.get_attribute('href'))
    WebDriverWait(browser,30).until(EC.presence_of_element_located((By.XPATH,"//a[@class='a-link-normal a-text-normal']")))
    ## getting URL's of first 10 laptops available on page
    urls=[]
    for i in browser.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")[0:10]:urls.append(i.get_attribute('href'))
    
    ## getting all required data
    titles=[]
    ratings=[]
    price=[]
    for i in urls:
        browser.get(i)
        browser.refresh()
        soup= BeautifulSoup(browser.page_source, 'html.parser')
        rat = soup.find('div', attrs = {'id':'averageCustomerReviews'})
        if rat is not None:ratings.append(rat.find('i').find('span').text)
        else: ratings.append('Null')
        try:
            price.append(browser.find_element_by_xpath("//td[@class='a-span12']/span[1]").text)
        except:
            price.append('Not available')
        titles.append(browser.find_element_by_id('productTitle').text)
 
    ## comparing the lenght of lists   
    if((len(titles)==len(ratings)) & (len(ratings)==len(price)) ):
        df=pd.DataFrame({'Title':titles,'Rating':ratings,'Price':price})
    else:
        print('Lenght of lists do not match')
    #browser.close()
    return df
```
