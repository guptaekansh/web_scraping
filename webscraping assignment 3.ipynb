```python
##### getting the URL of all products of first three pages
urls=[]
for k in range(3):
    for i in browser.find_elements_by_xpath("//div[@class='a-section a-spacing-medium']//h2"):
        urls.append(i.find_element_by_xpath(".//a").get_attribute('href'))
    try:
        next_page=browser.find_element_by_xpath("//div[@class='a-text-center']/ul/li[@class='a-last']/a").get_attribute('href')
        browser.get(next_page)
        browser.refresh()   
    except:
        break
## declaring empty lists to collect data
brand_name=[]
name=[]
rating=[]
no_of_rating=[]
price=[]
exp_delivery=[]
return_exchange=[]
other_details=[]
availability=[]
url=[]

## traversing all URL one by one
for i in urls:
    browser.get(i)
    url.append(i)
    ## getting brand_name
    try: brand_name.append(browser.find_element_by_id("productTitle").text.split(' ')[0]) 
    except: brand_name.append('-')
    ## getting name of product
    try: 
        st=''
        for i in browser.find_element_by_id("productTitle").text.split(' ')[1:]:
            st=st+' '+i
        name.append(st)
    except: name.append('-')
    ## getting rating of the product
    try: rating.append(browser.find_element_by_id("acrPopover").get_attribute('title')) 
    except: rating.append('-')
    ## getting number of ratings
    try: no_of_rating.append(browser.find_element_by_id("acrCustomerReviewText").text) 
    except: no_of_rating.append('-')
    ## getting price
    try: 
        try: price.append(browser.find_element_by_id("priceblock_saleprice").text) 
        except:
            try: price.append(browser.find_element_by_id("priceblock_dealprice").text)
            except: price.append(browser.find_element_by_id("priceblock_ourprice").text)
    except: price.append('-')
    ## return/exchnage
    try: 
        row=browser.find_element_by_xpath("//div[@class='a-row icon-farm-wrapper']").find_elements_by_xpath(".//div")
        found=False
        for i in row:
            if(i.get_attribute('data-name')=='RETURNS_POLICY'):
                found=True
                return_exchange.append(i.find_element_by_xpath(".//span[1]/div[2]/a").text)
        if(found==False):
            return_exchange.append('-')
    except: return_exchange.append('-')
    ## expected_delivery
    try: exp_delivery.append(browser.find_element_by_id("ddmDeliveryMessage").find_element_by_xpath(".//b").text)
    except: exp_delivery.append('-')
    ## availablity
    try: 
        try: availability.append(browser.find_element_by_id("availability").find_element_by_xpath(".//span").text)
        except: availability.append(browser.find_element_by_id("deal_availability").find_element_by_xpath(".//div/span").text)
    except: availability.append('-')
    ## other_details
    try:
        details=[i.text.replace('\n','---') for i in browser.find_element_by_id("productDetails_techSpec_section_1").find_elements_by_xpath(".//tbody")] 
        other_details.append(details[0])
    except: other_details.append('-')
## framing a dataframe
df=pd.DataFrame({'Brand_name':brand_name,'Name of Product':name,'Rating':rating,'No.of ratings':no_of_rating,'Price':price,
             'Return/Exchange':return_exchange,'Expected Delivery':exp_delivery,'Availability':availability,'Other_details':other_details,
             'Product_URL':url})
## exporting dataframe as csv
df.to_csv('products.csv')
```

```python
def solution_3(URL):
    
    browser=webdriver.Chrome('chromedriver.exe')
    ## maximizing the chrome window
    browser.maximize_window()
    
    #^^^^^^^^^^^^ making one inner function to scrap the data accroding to keywords given in it as parameter^^^^^^^^^^^^^^^^
    def get_url(keywords):
        browser.get(URL)
        # finding the search box
        search_box=browser.find_element_by_xpath("//input[@class='gLFyf gsfi']")
        # sending data to search box
        search_box.send_keys(keywords)
        # finding search button
        search_button=browser.find_element_by_xpath("//button[@class='Tg7LZd']")
        # clicking search button
        search_button.click()
        
        # adding some delay to allow page to get loaded fully
        time.sleep(10)
        
        ## scrolling some pages so that the images is clearly visible (gets loaded) and do not cause any error while scraping
        for k in range(4):
            browser.find_element_by_xpath("//a[@class='wXeWr islib nfEiy mM5pbd']").send_keys(Keys.PAGE_DOWN)
        ## again adding some delay to allow images to be clearly visible after scrolling
        time.sleep(5)
        anchors=[img for img in browser.find_elements_by_xpath("//a[@class='wXeWr islib nfEiy mM5pbd']")[:100]]
        url=[]
        print(len(anchors))
        if(len(anchors)==100):
            for i in anchors:
                try:
                    i.click()
                    url.append(i.get_attribute('href'))
                ## in case image in not interactable append 'None' in that case
                except:
                    url.append('None')
            return url
        else:
            
            print('The page might not have been loaded properly due to which all images are not captured')
            
    #^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^ inner function ends ^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
    
    ######################################### calling inner function one by one###########################################
    #***************scraping data for Fruits******************
    fruits=get_url('fruits')
    #***************scraping data for cars******************
    cars=get_url('cars')
    #***************scraping data for Machine Learning******************
    ML=get_url('Machine Learning')
    if((len(fruits)==len(cars)) & (len(cars)==len(ML))):
        df=pd.DataFrame({'Fruits':fruits,'cars':cars,'ML':ML})
    else:
        print('Lenght of lists do not match')
    return df
```

## Solution 4

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
```

```python
def solution_4(URL,smartphone_name):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.maximize_window()
    browser.get(URL)
    ## searching for input field and entering smartphone name
    browser.find_element_by_xpath("//div[@class='_3OO5Xc']/input").send_keys(smartphone_name)
    time.sleep(1)
    ## clicking search button
    browser.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
    ## waiting till page gets loaded
    WebDriverWait(browser,60).until(EC.presence_of_element_located((By.XPATH,"//div[@class='_2kHMtA']/a[@class='_1fQZEK']")))
    ## storing all page URL's
    urls=[i.get_attribute('href') for i in browser.find_elements_by_xpath("//div[@class='_2kHMtA']/a[@class='_1fQZEK']")]
    ## declaring empty lists
    brand=[]
    name=[]
    color=[]
    RAM=[]
    ROM=[]
    p_camera=[]
    s_camera=[]
    display=[]
    d_resolution=[]
    processor=[]
    processor_core=[]
    battery=[]
    price=[]
    counter=0
    ## traversing all URL's
    for url in urls:
        browser.get(url)
        brand.append(browser.find_element_by_xpath("//span[@class='B_NuCI']").text.split(' ')[0])
        browser.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()
        ## storing all labels of data specification
        labels=browser.find_elements_by_xpath("//tr[@class='_1s_Smc row']/td[@class='_1hKmbr col col-3-12']")
        ## storing all data against each data specification
        data=browser.find_elements_by_xpath("//tr[@class='_1s_Smc row']/td[@class='URwL2w col col-9-12']")
        ## traversing each label and its data and storing data in lists
        for i,k in zip(labels,data):
            if(i.text=='Model Name'):
                name.append(k.text)
            if(i.text=='Color'):
                color.append(k.text)
            if(i.text=='RAM'):
                RAM.append(k.text)
            if(i.text=='Internal Storage'):
                ROM.append(k.text)
            if(i.text=='Primary Camera'):
                p_camera.append(k.text)
            if(i.text=='Secondary Camera'):
                s_camera.append(k.text)
            if(i.text=='Display Size'):
                display.append(k.text)
            if(i.text=='Resolution'):
                d_resolution.append(k.text)
            if(i.text=='Processor Type'):
                processor.append(k.text)
            if(i.text=='Processor Core'):
                processor_core.append(k.text)
            if(i.text=='Battery Capacity'):
                battery.append(k.text)
        counter=counter+1
        #### if any data is not availble then appending '-' to match the length of list
        if(len(name)!=counter):name.append('-')
        if(len(color)!=counter):color.append('-')
        if(len(RAM)!=counter):RAM.append('-')
        if(len(ROM)!=counter):ROM.append('-')
        if(len(p_camera)!=counter):p_camera.append('-')
        if(len(s_camera)!=counter):s_camera.append('-')
        if(len(display)!=counter):display.append('-')
        if(len(d_resolution)!=counter):d_resolution.append('-')
        if(len(processor)!=counter):processor.append('-')
        if(len(processor_core)!=counter):processor_core.append('-')
        if(len(battery)!=counter):battery.append('-')
        price.append(browser.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']").text)
        
    df=pd.DataFrame({'Brand_Name':brand,'Smartphone name':name,'Color':color,'RAM':RAM,'ROM':ROM,'primary camera':p_camera,
             'secondary camera':s_camera,'Display size':display,'Resolution':d_resolution,'Processor':processor,
             'Processor cores':processor_core,'Battery capacity':battery,'price':price,'URL':urls})
    df.to_csv('mobile_data.csv')
    return df
```

## Solution 7

```python
def solution_7(URL):
    browser=webdriver.Chrome('chromedriver.exe')
    browser.maximize_window()
    browser.get(URL)
    #>>>>>>>>>>>>>>>>>>>> getting names of laptops  <<<<<<<<<<<<<<<<<<<<<<<<
    names=[i.text for i in browser.find_elements_by_xpath("//a/h3")]
    #>>>>>>>>>>>>>>>>>>>> getting description of laptops  <<<<<<<<<<<<<<<<<<<<<<<<
    descriptions=[i.text for i in browser.find_elements_by_xpath("//div[@class='Section-center']/p")]
    ###### declaring empty lists to collect data####
    OS=[]
    display=[]
    processor=[]
    memory=[]
    weight=[]
    dimension=[]
    graphics=[]
    ctr=-1  ## counter
    price=[]
    
    ###### getting main html tags for all 10 laptops #####
    mains=browser.find_elements_by_xpath("//div[@class='Section-center']")
    spec_rows=[]
    for i in mains:
        ##### getting the tags for table body of each main tags #######
        spec_rows.append(i.find_element_by_xpath(".//div[@class='Spcs-details']/table/tbody"))
    for i in spec_rows:
        ctr=ctr+1  ##### taking count of each laptop data
        try: ### try to get price if it exists in normal rows along with other rows
            price.append(i.find_element_by_xpath(".//tr[10]/td[3]").text)
        except:  #### if price not found in normal rows then search for bold rows using below address of main tag
            price_rows=mains[ctr].find_elements_by_xpath(".//div[@class='Top10Price-Wrap']/div[@class='merchantdatalist1']/table/tbody/tr")
            st='' ## deaclaing empty string to append all available prices in bold tags
            for x in price_rows:## traversing all available  price rows other than normal table rows
                st=st+(x.find_element_by_xpath(".//td[3]/strong").text)+' '
            price.append(st) 
        keys=i.find_elements_by_xpath(".//tr") ## storing all table rows inside table body
        for k in keys:## traversing all rows inside each available table body
            #### storing data
            try:
                if(k.find_element_by_xpath(".//td[1]").text=='OS'):
                    OS.append(k.find_element_by_xpath(".//td[3]").text)
                if(k.find_element_by_xpath(".//td[1]").text=='Display'):
                    display.append(k.find_element_by_xpath(".//td[3]").text)
                if(k.find_element_by_xpath(".//td[1]").text=='Processor'):
                    processor.append(k.find_element_by_xpath(".//td[3]").text)
                if(k.find_element_by_xpath(".//td[1]").text=='Memory'):
                    memory.append(k.find_element_by_xpath(".//td[3]").text)
                if(k.find_element_by_xpath(".//td[1]").text=='Weight'):
                    weight.append(k.find_element_by_xpath(".//td[3]").text)
                if(k.find_element_by_xpath(".//td[1]").text=='Dimension'):
                    dimension.append(k.find_element_by_xpath(".//td[3]").text)
                if(k.find_element_by_xpath(".//td[1]").text=='Graphics Processor'):
                    graphics.append(k.find_element_by_xpath(".//td[3]").text)
            except:
                continue
    df=pd.DataFrame({'Name':names,'Description':descriptions,'OS':OS,'Display':display,'Processor':processor,'Memory':memory,
                    'Weight':weight,'Dimension':dimension,'Graphics':graphics,'Price':price})
    return df
```
